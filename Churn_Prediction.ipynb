```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Churn_Modelling.csv")
```

```python
data.head()
```

```python
X = data.iloc[:, 3:-1].values
Y = data.iloc[:, -1].values
```

```python
X, Y
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
label_encoder_1 = LabelEncoder()
X[:, 1] = label_encoder_1.fit_transform(X[:, 1])
label_encoder_2 = LabelEncoder()
X[:, 2] = label_encoder_2.fit_transform(X[:, 2])
```

```python
X
```

```python
onehotencoder = OneHotEncoder(categorical_features=[1])
X = onehotencoder.fit_transform(X).toarray()
```

```python
X = X[:, 1:]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
def build_classifier(optimizer):
    classifier = Sequential()
    classifier.add(Dense(units=6, kernel_initializer='uniform', activation='relu', input_shape=(11,)))
    classifier.add(Dense(units=6, kernel_initializer='uniform', activation='relu'))
    classifier.add(Dense(units=6, kernel_initializer='uniform', activation='relu'))
    classifier.add(Dense(units=1, kernel_initializer='uniform', activation='sigmoid'))
    classifier.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
    return classifier
```

```python
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV
```

```python
classifier = KerasClassifier(build_fn=build_classifier)
parameters = {'batch_size':  [16, 32],
             'epochs': [25, 50], 
             'optimizer': ['Adam', 'Adadelta']}
grid_search = GridSearchCV(estimator=classifier, param_grid=parameters)
```

```python
grid_search.fit(X_train, Y_train)
```

```python
best_param = grid_search.best_params_
best_score = grid_search.best_score_
```

```python
print("Best parameters {} - Best score {}".format(best_param, best_score))
```
